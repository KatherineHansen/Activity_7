# Violinplot

<img src="./Images/plots.png" alt="Drawing" style="width: 550px;"/>

>~Mixture of a boxplot and 2 KDE(kernel density estimation) plots.<br>
~Similar to boxplots in that it can show median and IQR, and give a sense of wiskers. <br>
~Differences: Does not show outliers and has a KDE.

Data:

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.head()
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4,6,8])] #only cars with even cylinders
```

Plot:

```python
sns.set_style('whitegrid')
sns.violinplot(cars.displacement);
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement); # Specify axes
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin', data=cars); # hue splits apart further
```

Symmetry is just aesthetics

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin', data=cars[cars.origin.isin(['japan', 'europe'])]) 
plt.legend(loc=2);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin', data=cars[cars.origin.isin(['japan', 'europe'])]
              ,split=True  # split needs exactly 2 categories for hue
              ) 
plt.legend(loc=2);
```

Seperate info from box

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin', data=cars[cars.origin.isin(['japan', 'europe'])]
              ,split=True
              ,inner='quartile'  # inner changes how datapoint info is represented
              ) 
plt.legend(loc=2);
```

# Scaling
Area of both sides are the same by default

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin', data=cars[cars.origin.isin(['japan', 'europe'])]
              ,split=True
              ,scale='count'  # First splits by cylinders then by count
              ) 
plt.legend(loc=2);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin', data=cars[cars.origin.isin(['japan', 'europe'])]
              ,split=True
              ,scale='count'  # splits then counts
              ,scale_hue=False  # turns that off
              ) 
plt.legend(loc=2);
```

```python
(cars[cars.origin.isin(['japan', 'europe'])]
    .groupby('cylinders')
    .origin
    .value_counts()
)
```

# Styling

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
              order=['japan', 'europe', 'usa']);
# List of order you want. Not in list, not shown.
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement,
              linewidth=5); # Width of grey lines
```

```python
sns.violinplot(x=cars.cylinders, y= cars.displacement,
              bw=.2); # the  kernel bandwidth
```
